```python
#Import requests for web scraping
import pandas as pd
import requests as rq
import numpy as np
import collections

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

import matplotlib as mpl
import matplotlib.pyplot as plt

# import k-means from clustering stage
from PIL import Image 
from matplotlib import rcParams

print('Libraries installed')
```

```python
website_url= rq.get('https://movimientociudadano.mx/noticias/me-vas-conocer-por-hacer-lo-correcto-discurso-de-luis-donaldo-colosio').text #Bring the data from the target URL
```

```python
#Import BeautifulSoup for html structure information from our request
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')

data = [element.text for element in soup.find_all("p")] #Retrieving text from "p" founds
data = str(data) #Convert to string
data
```

```python
from stop_words import get_stop_words #The words we want to ignore

stop_words_es = get_stop_words('es') #Spanish StopWords
```

```python
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
stop_words = ["año", "mil", "millones", "pesos", "ciento", "El", "En"] + stop_words_es #Adding aditional StopWords
#initiate world cloud object
colosio_ws = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stop_words
)

# generate the word cloud
colosio_ws.generate(data)
```

```python
# display the word cloud
plt.imshow(colosio_ws, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# re-generate the word cloud
colosio_ws.generate(data)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(colosio_ws, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
!wget --quiet http://politiks360.com/wp-content/uploads/2019/04/Copia-de-image_2019-04-08_11-01-50.png

mx_mask = np.array(Image.open('Copia-de-image_2019-04-08_11-01-50.png'))
    
print('Image downloaded and saved!')
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(mx_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object
colosio_ws = WordCloud(background_color='white', max_words=2000, mask=mx_mask, stopwords=stop_words)

# generate the word cloud
colosio_ws.generate(data)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height


plt.imshow(colosio_ws, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
filtered_words = [word for word in data.split() if word not in stop_words]
counted_words = collections.Counter(filtered_words)

words = []
counts = []
for letter, count in counted_words.most_common(10):
    words.append(letter)
    counts.append(count)
```

```python
colors = cm.rainbow(np.linspace(0, 1, 10))
rcParams['figure.figsize'] = 20, 10

plt.title('Top words in the text vs their count')
plt.xlabel('Count')
plt.ylabel('Words')
plt.barh(words, counts, color=colors)
```
